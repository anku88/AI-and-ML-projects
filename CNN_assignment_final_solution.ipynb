```python
from sklearn.utils import shuffle
X,Y = shuffle(X,Y,random_state=4)
```

***Splitting the data*** **into** **test** **and** **train**

```python
from sklearn.model_selection import train_test_split
X_train,X_val,Y_train,Y_val = train_test_split(X,Y,test_size = 0.20)
```

```python
print(X_train.shape,X_val.shape)
print(Y_train.shape,Y_val.shape)
```

###### **Supervised Learning algorithm (Logistic Regression) for model training**

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression(random_state=10)
reshaped_x_train = X_train.reshape(X_train.shape[0], -1)
modelLR.fit(reshaped_x_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
reshaped_x_val = X_val.reshape(X_val.shape[0], -1)
LR_preds = modelLR.predict(reshaped_x_val)
LRscore = accuracy_score(Y_val,LR_preds)
print('Accuracy Score of Logistic Reression model is :',LRscore)
```

```python
GNB_preds = modelGNB.predict(reshaped_x_val)
GNBscore = accuracy_score(Y_val,GNB_preds)
print('Accuracy Score of GNB  model is :',GNBscore)
```

**DNN** **model**

```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, Reshape
from keras.layers import Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint, EarlyStopping
#Installing Tensorboard for Colab
!pip install tensorboardcolab
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
Y = pd.DataFrame(Y)
encoder = LabelEncoder()
encoder.fit(Y[0])
encoded_labels = encoder.transform(Y[0])
categorical_labels = tf.keras.utils.to_categorical(encoded_labels)
```

```python
x_train,x_val,y_train,y_val=train_test_split(X,categorical_labels,test_size=0.3)
```

```python


#Clear out model from current memory
tf.keras.backend.clear_session()

#Initialize Sequential model
modelDNN = Sequential()

#Reshape data from 2D to 1D -> 128x128x3 to 49152
modelDNN.add(Reshape((49152,),input_shape=(128,128,3)))

#Normalize the data
modelDNN.add(BatchNormalization())

#Hidden layers with batch normalization
#Layer 1
modelDNN.add(Dense(256, activation = 'relu'))
modelDNN.add(BatchNormalization())
#Layer 2
modelDNN.add(Dense(128, activation ='relu'))
modelDNN.add(BatchNormalization())
modelDNN.add(Dropout(0.5))
#Layer 3
modelDNN.add(Dense(64 , activation ='relu'))
modelDNN.add(BatchNormalization())
modelDNN.add(Dropout(0.3))

#Output layer contains 12 nodes for 12 classes
modelDNN.add(Dense(12, activation ='softmax'))
```

```python
modelDNN.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Model Summary
modelDNN.summary()
```

```python
model_checkpoint=tf.keras.callbacks.ModelCheckpoint("/content/drive/MyDrive/CNNProject1/plantspecies_modelDNN.h5",
                                                    save_best_only=True, 
                                                    monitor='val_accuracy', 
                                                    mode='max', 
                                                    verbose=1)
```

```python
train_acc = history_DNN.history['accuracy']
val_acc = history_DNN.history['val_accuracy']
print('best training_accuracy our model achieved is :',max(train_acc))
print('best test_accuracy our model achieved is:',max(val_acc))
```

**CNN**

```python
input_shape = (128, 128, 3)
num_classes = 12
```

```python
# Create model2
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Flatten,Dense,BatchNormalization,Dropout, Activation
from tensorflow.keras.layers import Conv2D, MaxPool2D,MaxPooling2D
from tensorflow.keras.optimizers import SGD, Adam

#Clear out model from current memory
tf.keras.backend.clear_session()

row = 128
col = 128
channels = 3
n_classes = 12
n_epochs = 50
batch_size = 32


#Initialize Sequential model
modelCNN = tf.keras.models.Sequential()

#Conveolution Layer
modelCNN.add(Conv2D(32, kernel_size= 3, activation= 'relu', input_shape = (row,col,channels)))

#MaxPool Layer
modelCNN.add(MaxPooling2D((2,2)))

#Dropout layer
modelCNN.add(tf.keras.layers.Dropout(0.2))

#Conveolution Layer
modelCNN.add(Conv2D(64, kernel_size= 3, activation= 'relu'))

#MaxPool Layer
modelCNN.add(MaxPooling2D((2,2)))



#Conveolution Layer
modelCNN.add(Conv2D(128, kernel_size= 3, activation= 'relu'))

#MaxPool Layer
modelCNN.add(MaxPooling2D((2,2)))


#Faltten before feeding into Dense layer
modelCNN.add(Flatten())

#Output layer contains 2 nodes for 2 classes
modelCNN.add(Dense(n_classes, activation ='softmax'))



```

```python
modelCNN.summary()
```

```python
from keras.optimizers import Adam
from keras.losses import categorical_crossentropy

#To use adam optimizer for learning weights with learning rate = 0.001
optimizer = Adam(lr=0.001)
#Set the loss function and optimizer for the model training
modelCNN.compile(loss=categorical_crossentropy,
              optimizer=optimizer,
              metrics=['accuracy'])
```

```python
#Adding Early stopping callback to the fit function is going to stop the training,
#if the val_loss is not going to change even '0.001' for more than 10 continous epochs

early_stopping = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=10)

# create checkpoint
model_checkpoint = tf.keras.callbacks.ModelCheckpoint("/content/drive/MyDrive/CNNProject1/plantspecies_modelCNN2.h5", save_best_only=True, monitor='val_accuracy', mode='max', verbose=1)
```

```python
history_CNN = modelCNN.fit(x_train, y_train ,batch_size=32,epochs=20,verbose=1,validation_data=(x_val, y_val),callbacks=[model_checkpoint])
```

## **Possible reasons behind these observation **
1. ANN/DNNs  model can handle  non-linear and complex relationships better as compared to classical supervised learning models.It works on pixel intesity distribution and provides more accuracy but at the same time requires more and more data.Both ML model and ANN model  with large amount of image data likely to have more run time but ML model underperforms as far as accuracy is concerned because it does not looks for pixel intesity distribution.
